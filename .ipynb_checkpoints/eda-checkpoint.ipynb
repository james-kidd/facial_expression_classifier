# Create DataFrame

## Load CSV file to dataframe.
The csv file downloaded from kaggle contains only 3 columns: 'emotion' (int), 'usage' (str), and 'pixels' (str).

## Clean and re-organize dataframe
Run the face_csv_to_df.py file to convert the 'pixels' column to an appropriate format (np.array), and proper array 
shape in order to match the original image dimensions; (48,48).

The 'emotion' column contains emotion labels as a interger type (ranging for 0 to 6) for the pixel data given in a row. Due to this, 
an additional column 'label_name' (str) will be added to the dataframe to indicate the emotion corresponding to the 'emotion' column. 
The newly created column will later be used in the 'creat_img_dir' function.

Furthermore, rows containing a 'usage' instance of the values 'PrivateTest' and 'PublicTest' are changed to 'Test'. 


```python
#directories
face_csv_filepath = 'data/icml_face_data.csv'

#load csv as pandas dataframe with proper values and column names.
face = face_csv_to_df(face_csv_filepath)
```

## Delete corrupted and duplicate images from dataframe

From initial inspection of data, certain images within csv were found to be corrupt (blank image).
In addition, images were found to be duplicated multiple times with different emotion labels attached to the image.

To solve this issue, corrupted images were simply removed from dataframe. To solve the issues of duplicates, all but one duplicates were removed 
from the dataframe. Pixel data was used to find duplicates, and the image with a emotion label considered to be the most scarce was kept.

```python
#delete corrupted and duplicate images from dataframe
face = delete_duplicate_img(face)
```
